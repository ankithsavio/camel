```python
from hybrid_browser_debug import BrowserDebugDemo

demo = BrowserDebugDemo(headless=False, cache_dir="debug_output")

result = await demo.start()
```

```python
async def get_pw_count(ref: str):
    page = await demo.toolkit._require_page()

    selector = f'[aria-ref="{ref}"]'
    count = await page.locator(selector).count()

    print(count)
```

```python
import re

async def test_click(ref: str):
    result = await demo._cmd_click([ref])
    if "snapshot not changed" in result:
        print("Click didnt trigger")
```

```python
wikipedia_pattern = r"^.*wikipedia.*$"
link_pattern = r"^.*link .*$"
```
